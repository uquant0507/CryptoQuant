```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from google.colab import drive
drive.mount('/content/drive')
```

```python
class Factor:
  def __init__(self, data):
    self.data = data.dropna()
    self.pct_change = self.get_pct_change(self.data)
    self.momentum = self.get_momentum(self.pct_change, 1)
    self.result = self.analyze(self.pct_change, self.momentum)
    print(self.result)
  
  def get_pct_change(self, prices, n=1):
    prices = prices.copy(deep=True)
    pct_change = prices.pct_change(periods=n).dropna()
    return pct_change

  def get_momentum(self, pct_change, n=1):
    momentum = pct_change.shift(n).dropna()
    return momentum
  
  def analyze(self, prices, factor1, factor2 = None):
    if factor2 == None:
      factor1_coef = []
      market_coef = []
      prices = prices.copy(deep=True)
      factor1 = factor1.copy(deep=True)
      for _ in range(len(prices) - len(factor1)):
        prices = prices.drop(prices.index[0])
      for i in range(factor1.shape[0]):
        reg = LinearRegression(fit_intercept=True)
        X = factor1[i:i+1].transpose()
        Y = prices[i:i+1].transpose()
        reg.fit(X, Y)
        factor1_coef.append(reg.coef_[0][0])
        a = reg.coef_.shape
      factor1_coef = np.array(factor1_coef)
      market_coef = np.array(market_coef)
      t_stat = np.sqrt(len(factor1_coef)) * factor1_coef.mean()/ factor1_coef.std()
      return t_stat

```
